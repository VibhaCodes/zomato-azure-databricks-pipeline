```python
df_gold = spark.read.format("delta") \
    .load("abfss://zomato-external@zomatostorage.dfs.core.windows.net/gold/top_restaurants_city")

display(df_gold)
```

```python
city_summary = spark.table("zomato_catalog.gold.city_summary")
top_restaurants = spark.table("zomato_catalog.gold.top_restaurants")
city_rating_summary = spark.table("zomato_catalog.gold.city_rating_summary")

display(city_summary)
display(top_restaurants)
display(city_rating_summary)
```

Top 10 restaurants

```python
%sql
SELECT *
FROM zomato_catalog.gold.top_restaurants
ORDER BY avg_rating DESC;
```

City-wise restaurant count

```python
%sql
SELECT *
FROM zomato_catalog.gold.city_summary
ORDER BY restaurant_count DESC;
```

City-wise average rating

```python
%sql
SELECT *
FROM zomato_catalog.gold.city_rating_summary
ORDER BY avg_rating DESC;
```
